```python
filename = 'rankings/1_to_50.txt'
```

```python
f = open(filename, 'r')
```

```python
def index_after_find(main_str, sub_str):
    i = main_str.find(sub_str)
    if i == -1: return i
    return i + len(sub_str)
    
```

```python
def format_country(country):
    if country == 'United States':
        return 'USA'
    elif 'China' in country:
        return 'China'
    elif 'Hong Kong' in country:
        return 'Hong Kong'
    elif country == 'United Kingdom':
        return 'UK'
    else:
        return country
```

```python
def remove_bracket(uni):
    if uni[-1] == ')':
        idx = uni.rindex('(')
        uni = uni[:idx].strip()
    return uni
```

```python
rankings_dict = {}
```

```python
def add_to_rankings(uni, country, rank):
    if country not in rankings_dict:
        rankings_dict[country] = {}
#         rankings_dict[country] = []

    rankings_dict[country][uni] = rank
#     rankings_dict[country].append((uni, rank))
```

```python
for line in f:
    if len(line) < 10:
        rank = line.strip()
        if rank[0] == '=': rank = rank[1:]
        continue
    else:
        uni = line[:line.find(' Logo')].strip()
        uni = remove_bracket(uni)
        country = line[index_after_find(line, 'More'):].strip()
        country = format_country(country)
        
        add_to_rankings(uni, country, rank)
```

## Lower Ranks
Universities are not assigned individual ranks, but rather range ranks such as 51-100 and 101-150.
We assign the average rank to these institutions (e.g. rank 51-100 is recorded as rank 75)

```python
filename = 'rankings/51-600.txt'
```

```python
f = open(filename, 'r')
```

```python
def format_rank(rank):
    sp = rank.split('-')
    if len(sp) < 2: return rank
    new_rank = str(int((int(sp[0]) + int(sp[1]) - 1)/2))
    return new_rank
```

```python
for line in f:
    if len(line) < 10:
        rank = format_rank(line.strip())
        continue
    else:
        uni = line[:line.find(' Logo')].strip()
        uni = remove_bracket(uni)
        country = line[index_after_find(line, 'More'):].strip()
        country = format_country(country)
        
        add_to_rankings(uni, country, rank)
```
