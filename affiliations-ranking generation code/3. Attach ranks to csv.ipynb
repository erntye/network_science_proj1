```python
import pandas as pd
import json
```

```python
rank_file = 'rankings_dict.json'
affiliation_file = 'author_inst_aggregated.csv'
```

```python
with open(rank_file, 'r') as f:
    rank_dict = json.load(f)
```

```python
df = pd.read_csv(affiliation_file)
```

```python
df1 = df[:5000].copy()
```

```python
df1['rank'] = '1000'
```

```python
def add_rank(row):
    if row['affiliation'] == '-':
        return row
    split = row['affiliation'].split(',')
    country = split[-1].strip()
    uni = ','.join(split[:-1])
    if country in rank_dict and uni in rank_dict[country]:
        row['rank'] = rank_dict[country][uni]
    return row
```

```python
df1.apply(add_rank, axis=1)
```

```python
df1['rank'].unique()
```

```python
import cProfile
```

```python
def time_test(num):
    df1 = df[:num].copy()
    df1['rank'] = '1000'
    df1.apply(add_rank, axis=1)
```

```python
cProfile.run('time_test(100000)')
```
